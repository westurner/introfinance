```python
from pandas import DataFrame, Series, Index
import numpy.lib.financial as fin
from sympy import abc, Eq, symbols, solve
```

```python
# 2.2
# annuity: future value
idx = Index(range(0,4), name='year')
df = DataFrame(index=idx)
df['cash flow'] = Series(['0','C','C','C'])
df['years to end'] = Series([3,2,1,0])
df['future value'] = Series([0, 'C*((1+r)**2)', 'C*((1+r)**1)', 'C'])
display(df)
```

```python
# fv of annuity: formula
# fv = (C*((1+r)**2))+(C*((1+r)**1))+(C*((1+r)**0))
# fv = C*( (1+r)**2 + (1+r) + 1 )
# fv = C *( (1+r)**(n-1) + ... )
fva = abc.c * ( (1+abc.r)**(abc.n-1) )
display(fva)
print("+ ...")
```

```python
# example 1: fv of an annuity
# c == pmt
# what will be the value of your portfolio at retirement
# if you deposit 10000 every year in a pension fund?
# you plan to retire in 40 years and expect to earn 8%
rate=0.08
nper=40
pmt=10000

future_value = -1*fin.fv(rate, nper, pmt, 0)
display(future_value)
#fin.pmt(rate, nper, )

idx = Index(range(0,40))
df = DataFrame(index=idx)
df['deposit'] = 10000
df['cumsum'] = df['deposit'].cumsum()

df['fv'] = Series(
  [  pmt*((1+rate)**(i)) for i in idx ]
)
df['fv_cumsum'] = df['fv'].cumsum()

display(df.plot())
display(df)
checkdiff = (df['fv_cumsum'][idx[-1]] - future_value)
assert checkdiff <= 0.01
#fin.pmt(rate, nper, 10000*nper)
```

```python
# example 2: FV of an annuity
# suppose you want to guarantee yourself $500,000 when you retire
# 25 years from now.
# how much must you invest each year, starting at the end of this year
# if the interest rate is 8%
fv = 500000
nper=25
rate=0.08

pmt = -1*fin.pmt(rate, nper, 0, fv)
fv_check = -1*fin.fv(rate, nper, pmt, 0)
assert (fv - fv_check) <= 0.01

# chart it
idx = Index(range(0,nper))
df = DataFrame(index=idx)
df['pmt'] = pmt
df['pmt_cumsum'] = df['pmt'].cumsum()
df['fv'] = Series(
  [ -1*fin.fv(rate, i+1, pmt, 0) for i in idx ]
)
df['return'] = df['fv']-df['pmt_cumsum']
df['return_pct'] = (df['return'] / df['pmt_cumsum'])*1e2
display(df.plot())
display(df)
```

```python
# 2.5 annuities: present values
idx = Index(range(0,4), name='year')
df = DataFrame(index=idx)
df['cash flow'] = Series(['0','C','C','C'])
df['years to discount'] = Series([0,1,2,3])
df['future value'] = Series([0, 'C / ((1+r)**1)', 'C / ((1+r)**2)', 'C / ((1+r)**3)'])
display(df)
```

```python
# PV of an Annuity: formula
# pv_3 = (c / (1+r)) + (c / (1+r)**2) + (c / (1+r)**3)
#      = c *( (1/(1+r)) + (1/(1+r))**2 + (1/(1+r)**3) )
# pv_n = c *( sigma( (1/(1+r)**(1->n) ) )
```

```python
# PV of an annuity: example 1
# how much money do you need in the bank today so that you can spend
# $10,000 every year for the next 25 years,
# starting at the end of this year
# suppose r = 5%
nper = 25
rate = 0.05
pmt = 10000

present_value = -1*fin.pv(rate, nper, pmt)
#present_value = fin.pv(0.05, 25, 0, 500000)
display(present_value)

idx = Index(range(0,nper+1))
df = DataFrame(index=idx)

df['initial'] = present_value
df['pmt'] = pmt
df['pmt'][0] = 0
df['pmt_cumsum'] = df['pmt'].cumsum()

def calculate_balance(initial, rate, nper, pmt):
    balance = initial
    yield balance
    _rate = 1+rate
    for i in xrange(0,nper):
        balance = balance * (_rate)
        balance = balance - pmt
        yield balance
    return

df['balance'] = Series(
   list( calculate_balance(present_value, rate, nper, pmt) )
)
df['interest'] = df['balance'].shift(1)*rate
df['interest_cumsum'] = df['interest'].cumsum()
df['payout'] = df['interest_cumsum'] + df['pmt_cumsum'].shift(1)
df['delta %'] = (df['payout'] / df['initial'])*100

display('interest: %f' % df['interest'].sum())

#df['total delta'] = df['balance'] - df['initial']
display(df.plot(figsize=(10,6)))
display(df)

df2 = DataFrame(index=idx)
df2['delta %'] = df['delta %']
df2['100'] = 100
df2['200'] = 200
display(df2.plot(figsize=(10,6)))
```

```python
# 2.6
# +++
# PV of an annuity: example 2
# you plan to attend business school and you will be forced to
# take out a $100,000 loan at 10%
# you want to figure out your yearly payments,
# given that you will have 5 years to pay back the loan
loan = 100000
rate = 0.10
nper = 5


pmt = fin.pmt(rate, nper, loan)
display(pmt)

fv = fin.fv(rate, nper, 0, loan)
display(fv)
fv_check = -1*fin.fv(rate, nper, pmt, 0)
display(fv_check)
assert (fv - fv_check) <= 0.01
```

```python
# 2.7
# +++
# a loan: the power of finance
balance = 100000
rate = 0.10
nper = 5

pmt = -1*fin.pmt(rate, nper, balance)
display(pmt)

def calculate_loan_amortization(rate, nper, balance, pmt):
    for per in xrange(0, nper):
        interest = rate*balance
        principal = pmt-interest
        yield (per, balance, pmt, interest, principal)
        balance = balance - principal
    return

data = calculate_loan_amortization(rate, nper, balance, pmt) 
idx = Index(range(nper))
df = DataFrame(data=list(data),
    index=idx,
    columns=['period','balance','pmt', 'interest', 'principal'])
del df['period']
display(df)
```
